```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

Step 4: create column for busiest period of the day.

```python
# check which hours of the day are the busiest
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

for this step, I will use the busiest and slowest 6 hours of the day, with the remaining 12 hours being regularly busy.

the busiest periods of time are 10-11, 14-15 and 12-13 (i.e. from 10-15). the slowest periods are 3-4, 2, 5 and 0-1 (i.e. from 0-5)

```python
#use for-loop to mark these periods of time in the df
result3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10,11,12,13,14,15]:
    result3.append("Most orders")
  elif value in [0,1,2,3,4,5]:
    result3.append("Fewest orders")
  else:
    result3.append("Average orders")
```

```python
result3
```

```python
df_ords_prods_merge['busiest_period_of_day']=result3
```
