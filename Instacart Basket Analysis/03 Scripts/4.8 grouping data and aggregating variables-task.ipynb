Creating a loyalty flag for existing customers

```python
#aggregate data using transform() function
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#check output of data aggregation
ords_prods_merge.head(15)
```

```python
#create loyalty flag for existing customers using loc()
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#check updated output
ords_prods_merge.head(15)
```

Determine spending habits of each of the 3 categories of customer by checking the prices of products they typically purchase.

```python
#check product price statistics for each loyalty group in the dataframe by grouping aggregate price data by loyalty group
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
#the max price value of 99999.0 appears to be an outlier, possibly an erroneous entry
```

Create a flag for different levels of spending based on average prices across orders

```python
#aggregate data using transform()
ords_prods_merge['avg_prod_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#create spending flag
ords_prods_merge.loc[ords_prods_merge['avg_prod_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_prod_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#check results for flag
ords_prods_merge[['user_id', 'spending_flag', 'avg_prod_price']].head(15)
```

```python
#check spending statistics for each customer category
ords_prods_merge.groupby('spending_flag').agg({'prices': ['min','mean', 'max']})
```

Determine frequent vs non-frequent customers

```python
#aggregate data by order frequency using transform()
ords_prods_merge['avg_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check for new columns
ords_prods_merge[['user_id', 'days_since_prior_order', 'avg_days_since_prior_order']].head(15)
```

```python
#create flag for average order frequency
ords_prods_merge.loc[ords_prods_merge['avg_days_since_prior_order'] > 20, 'orderfreq_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['avg_days_since_prior_order'] <= 20) & (ords_prods_merge['avg_days_since_prior_order'] > 10), 'orderfreq_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['avg_days_since_prior_order'] <= 10, 'orderfreq_flag'] = 'Frequent customer'
```

```python
#check for flag results
ords_prods_merge[['user_id', 'orderfreq_flag', 'avg_days_since_prior_order']].head(15)
```

```python
#check descriptive stats for flagged groups
ords_prods_merge.groupby('orderfreq_flag').agg({'days_since_prior_order': ['min','mean', 'median', 'max']})
```

```python
#get rid of outliers in prices
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
```
